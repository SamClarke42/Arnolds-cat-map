```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline
```

```python
def arnold_cat_map(image,iterations):
    x,y,z = image.shape
    #n = x
    image_array = np.zeros([x,y,z])
    for i in range(x):
        for j in range(y):
            image_array[i][j] = image[((2*i)+j)%x][(i+j)%x]
    plt.imsave("pic_iterate",image_array)
    return image_array
```

```python
pic = mpimg.imread('pic1.png')
for i in range(20):
    pic = arnold_cat_map(pic,i)
    plt.imsave("pic_iterate"+str(i),pic)
#26x26 image returns at i=42
#google image returns at 299, at 51 its small versions, at 59,119,179,239 its inverted colour
plt.imshow(pic)
```
