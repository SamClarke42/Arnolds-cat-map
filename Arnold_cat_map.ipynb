```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline
```

```python
def arnold_cat_map(image,iterations):
    
    x,y,z = image.shape     #Extract values of the dimensions of image
    image_array = np.zeros([x,y,z])     #Create matrix of 0s in size of image
    
    for i in range(x):
        for j in range(y):
            
            image_array[i][j] = image[((2*i)+j)%x][(i+j)%x]     #Apply the mapping of the image to the 0 matrix image_array
            
    plt.imsave("pic_iterate",image_array)     #Save image picture
    
    return image_array     #Function outputs the mapped matrix.
```

```python
pic = mpimg.imread('pic1.png')     #Read in image

for i in range(60):
    pic = arnold_cat_map(pic,i)     #Apply mapping function for i iterations
    plt.imsave("pic_iterate"+str(i),pic)     #Save image at each iteration.... ( +str(i) numbers the images). 
    
#26x26 image returns at i=42.
#google image returns at i=299.

plt.imshow(pic)     #Shows the final iteration image inline
```
